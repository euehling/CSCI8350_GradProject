# Paddy Yield Star Schema
## Week 4 — Fact & Dimension Tables with Synthetic Data

This notebook implements a star schema for the paddy yield dataset. The schema consists of one fact table and four dimension tables, populated with synthetic data that reflects realistic values from the original dataset.

### Schema Overview
- **Fact_Yield** — central fact table recording yield observations
- **Dim_Location** — geographic information (Agriblock)
- **Dim_Farm** — farm characteristics (soil type, nursery type)
- **Dim_Crop** — crop details (variety, fertilizer usage)
- **Dim_Weather** — weather conditions during growing season

```python
import pandas as pd
import numpy as np

np.random.seed(42)
```

## Step 1: Create Dimension Tables

Dimension tables provide descriptive context for the fact table. Each has a surrogate key used as a foreign key in Fact_Yield.

```python
# ── Dim_Location ─────────────────────────────────────────────────────────────
# Geographic segmentation based on Agriblock regions in the dataset

dim_location = pd.DataFrame({
    'location_key': [1, 2, 3, 4, 5, 6],
    'agriblock': ['Cuddalore', 'Kurinjipadi', 'Panruti', 'Kallakurichi', 'Sankarapuram', 'Chinnasalem'],
    'region': ['Cuddalore District', 'Cuddalore District', 'Cuddalore District',
                'Villupuram District', 'Villupuram District', 'Villupuram District']
})

print("Dim_Location")
print(dim_location)
print(f"Rows: {len(dim_location)}")
```

```python
# ── Dim_Farm ──────────────────────────────────────────────────────────────────
# Farm-level characteristics including soil type and nursery method

dim_farm = pd.DataFrame({
    'farm_key': [1, 2, 3, 4],
    'soil_type': ['clay', 'clay', 'alluvial', 'alluvial'],
    'nursery_type': ['dry', 'wet', 'dry', 'wet']
})

print("Dim_Farm")
print(dim_farm)
print(f"Rows: {len(dim_farm)}")
```

```python
# ── Dim_Crop ──────────────────────────────────────────────────────────────────
# Crop variety and fertilizer application details
# DAP applied at 20 days, Urea at 40 days, Potash at 50 days

np.random.seed(42)
n_crop = 20

dim_crop = pd.DataFrame({
    'crop_key': range(1, n_crop + 1),
    'variety': np.random.choice(['CO_43', 'ponmani', 'delux ponni'], n_crop),
    'dap_20days': np.random.randint(40, 241, n_crop),
    'urea_40days': np.round(np.random.uniform(27.13, 162.78, n_crop), 2),
    'potash_50days': np.random.randint(20, 120, n_crop),
    'micronutrients_70days': np.random.randint(5, 50, n_crop)
})

print("Dim_Crop (first 5 rows)")
print(dim_crop.head())
print(f"Rows: {len(dim_crop)}")
```

```python
# ── Dim_Weather ───────────────────────────────────────────────────────────────
# Weather conditions recorded across four 30-day growing periods
# Temperature ranges, humidity, wind speed and direction per period

np.random.seed(42)
n_weather = 30

wind_directions = ['SW', 'NW', 'ENE', 'W', 'SSE', 'E']

dim_weather = pd.DataFrame({
    'weather_key': range(1, n_weather + 1),

    # Period 1: Days 1-30
    'min_temp_d1_d30': np.round(np.random.uniform(18.0, 20.5, n_weather), 1),
    'max_temp_d1_d30': np.random.randint(31, 36, n_weather),
    'humidity_d1_d30': np.round(np.random.uniform(64.6, 88.5, n_weather), 1),
    'wind_speed_d1_d30': np.random.randint(4, 11, n_weather),
    'wind_dir_d1_d30': np.random.choice(wind_directions, n_weather),

    # Period 2: Days 31-60
    'min_temp_d31_d60': np.round(np.random.uniform(18.0, 20.5, n_weather), 1),
    'max_temp_d31_d60': np.random.randint(31, 36, n_weather),
    'humidity_d31_d60': np.round(np.random.uniform(64.6, 88.5, n_weather), 1),
    'wind_speed_d31_d60': np.random.randint(4, 11, n_weather),
    'wind_dir_d31_d60': np.random.choice(wind_directions, n_weather),

    # Period 3: Days 61-90
    'min_temp_d61_d90': np.round(np.random.uniform(18.0, 20.5, n_weather), 1),
    'max_temp_d61_d90': np.random.randint(31, 36, n_weather),
    'humidity_d61_d90': np.round(np.random.uniform(64.6, 88.5, n_weather), 1),
    'wind_speed_d61_d90': np.random.randint(4, 11, n_weather),
    'wind_dir_d61_d90': np.random.choice(wind_directions, n_weather),

    # Period 4: Days 91-120
    'min_temp_d91_d120': np.round(np.random.uniform(18.0, 20.5, n_weather), 1),
    'max_temp_d91_d120': np.random.randint(31, 36, n_weather),
    'humidity_d91_d120': np.round(np.random.uniform(64.6, 88.5, n_weather), 1),
    'wind_speed_d91_d120': np.random.randint(4, 11, n_weather),
    'wind_dir_d91_d120': np.random.choice(wind_directions, n_weather),
})

print("Dim_Weather (first 5 rows)")
print(dim_weather.head())
print(f"Rows: {len(dim_weather)}")
```

## Step 2: Create Fact Table

The fact table records individual yield observations and links to all four dimension tables via foreign keys. Contains 150 rows.

```python
# ── Fact_Yield ────────────────────────────────────────────────────────────────
# Central fact table recording yield per observation
# Foreign keys link to all four dimension tables

np.random.seed(42)
n_fact = 150

hectares = np.random.randint(1, 7, n_fact)
seed_rate = np.random.randint(25, 151, n_fact)
paddy_yield_kg = np.random.randint(5410, 38815, n_fact)
yield_per_hectare = np.round(paddy_yield_kg / hectares, 2)

fact_yield = pd.DataFrame({
    'yield_key': range(1, n_fact + 1),
    'location_key': np.random.randint(1, 7, n_fact),
    'farm_key': np.random.randint(1, 5, n_fact),
    'crop_key': np.random.randint(1, n_crop + 1, n_fact),
    'weather_key': np.random.randint(1, n_weather + 1, n_fact),
    'hectares': hectares,
    'seed_rate_kg': seed_rate,
    'paddy_yield_kg': paddy_yield_kg,
    'yield_per_hectare': yield_per_hectare
})

print("Fact_Yield (first 5 rows)")
print(fact_yield.head())
print(f"Rows: {len(fact_yield)}")
```

## Step 3: Export Synthetic Data to CSV

Export each table to CSV to simulate the data upload step. In a live Databricks environment, these CSVs would be uploaded via the Add Data UI or DBFS.

```python
# ── Export Tables to CSV ──────────────────────────────────────────────────────
# Simulates uploading synthetic data to Databricks

dim_location.to_csv('dim_location.csv', index=False)
dim_farm.to_csv('dim_farm.csv', index=False)
dim_crop.to_csv('dim_crop.csv', index=False)
dim_weather.to_csv('dim_weather.csv', index=False)
fact_yield.to_csv('fact_yield.csv', index=False)

print("CSV files exported:")
print("  dim_location.csv")
print("  dim_farm.csv")
print("  dim_crop.csv")
print("  dim_weather.csv")
print("  fact_yield.csv")
```

## Step 4: Load CSV Data and Insert into Tables

Reload each CSV and insert rows into the tables. This simulates the ETL load step in Databricks.

```python
# ── Load CSV Files ────────────────────────────────────────────────────────────
# Reload from CSV to simulate loading uploaded data into Databricks tables

dim_location_loaded  = pd.read_csv('dim_location.csv')
dim_farm_loaded      = pd.read_csv('dim_farm.csv')
dim_crop_loaded      = pd.read_csv('dim_crop.csv')
dim_weather_loaded   = pd.read_csv('dim_weather.csv')
fact_yield_loaded    = pd.read_csv('fact_yield.csv')

print("Tables loaded from CSV:")
print(f"  dim_location_loaded:  {len(dim_location_loaded)} rows")
print(f"  dim_farm_loaded:      {len(dim_farm_loaded)} rows")
print(f"  dim_crop_loaded:      {len(dim_crop_loaded)} rows")
print(f"  dim_weather_loaded:   {len(dim_weather_loaded)} rows")
print(f"  fact_yield_loaded:    {len(fact_yield_loaded)} rows")
```

```python
# ── INSERT Rows into Tables ───────────────────────────────────────────────────
# Simulate INSERT statements by appending loaded rows into in-memory tables
# In Databricks this would be: INSERT INTO table SELECT * FROM csv_view

# Initialize empty tables with correct schema
Dim_Location = pd.DataFrame(columns=dim_location_loaded.columns)
Dim_Farm = pd.DataFrame(columns=dim_farm_loaded.columns)
Dim_Crop = pd.DataFrame(columns=dim_crop_loaded.columns)
Dim_Weather = pd.DataFrame(columns=dim_weather_loaded.columns)
Fact_Yield = pd.DataFrame(columns=fact_yield_loaded.columns)

# INSERT — append all rows from loaded CSVs into each table
Dim_Location = pd.concat([Dim_Location, dim_location_loaded], ignore_index=True)
Dim_Farm = pd.concat([Dim_Farm, dim_farm_loaded], ignore_index=True)
Dim_Crop = pd.concat([Dim_Crop, dim_crop_loaded], ignore_index=True)
Dim_Weather = pd.concat([Dim_Weather, dim_weather_loaded], ignore_index=True)
Fact_Yield = pd.concat([Fact_Yield, fact_yield_loaded], ignore_index=True)

print("INSERT complete. Final row counts:")
print(f"  Dim_Location:  {len(Dim_Location)} rows")
print(f"  Dim_Farm:      {len(Dim_Farm)} rows")
print(f"  Dim_Crop:      {len(Dim_Crop)} rows")
print(f"  Dim_Weather:   {len(Dim_Weather)} rows")
print(f"  Fact_Yield:    {len(Fact_Yield)} rows")

print("\nSample INSERT — first 3 rows of Fact_Yield:")
print(Fact_Yield.head(3))
```

## Step 5: Validate Foreign Key Relationships

Verify that all foreign keys in Fact_Yield reference valid surrogate keys in the dimension tables.

```python
# ── Foreign Key Validation ────────────────────────────────────────────────────
# FK validation — keys were generated within valid ranges so this will always pass
# In a real ETL pipeline this would catch invalid references from external data sources

assert fact_yield['location_key'].isin(dim_location['location_key']).all(), "Invalid location_key found"
assert fact_yield['farm_key'].isin(dim_farm['farm_key']).all(), "Invalid farm_key found"
assert fact_yield['crop_key'].isin(dim_crop['crop_key']).all(), "Invalid crop_key found"
assert fact_yield['weather_key'].isin(dim_weather['weather_key']).all(), "Invalid weather_key found"

print("All foreign key relationships are valid.")
```

## Step 6: Schema Summary

```python
# ── Schema Summary ────────────────────────────────────────────────────────────

tables = {
    'Fact_Yield': fact_yield,
    'Dim_Location': dim_location,
    'Dim_Farm': dim_farm,
    'Dim_Crop': dim_crop,
    'Dim_Weather': dim_weather
}

print("=" * 50)
print("STAR SCHEMA SUMMARY")
print("=" * 50)
for table_name, df in tables.items():
    print(f"\n{table_name}")
    print(f"  Rows: {len(df)}")
    print(f"  Columns: {list(df.columns)}")
```

## Step 7: Sample Joined Query
Join the fact table to all dimension tables to show how the schema fits together.

```python
# ── Sample Join ───────────────────────────────────────────────────────────────
# Join fact table to all dimensions to produce a full denormalized view

joined = (fact_yield
    .merge(dim_location, on='location_key')
    .merge(dim_farm, on='farm_key')
    .merge(dim_crop[['crop_key', 'variety', 'dap_20days', 'urea_40days']], on='crop_key')
    .merge(dim_weather[['weather_key', 'min_temp_d1_d30', 'max_temp_d1_d30', 'humidity_d1_d30']], on='weather_key')
)

print("Joined View (first 5 rows)")
print(joined[['yield_key', 'agriblock', 'soil_type', 'nursery_type', 'variety',
              'hectares', 'yield_per_hectare', 'min_temp_d1_d30', 'humidity_d1_d30']].head())

print(f"\nAverage yield per hectare by Agriblock:")
print(joined.groupby('agriblock')['yield_per_hectare'].mean().round(2))
```
